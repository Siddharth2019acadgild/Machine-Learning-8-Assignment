```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
```

```python
text = soup.get_text()
```

```python
nltk.word_tokenize(text) # Preprocessing the text
```

```python
freq = nltk.FreqDist(nltk.word_tokenize(text)) # Applying freq distribution method
```

```python
from nltk import FreqDist
```

```python
import os
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
# Function to represent frequency analysis
def freq_words(text):
   
    
    fdist =FreqDist(text)
    words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())}) # Making a new datframe in the  
                                                                                       # form of Dictionary
     
    plt.figure(figsize=(20,15))
    ax = sns.barplot(data=words_df, x= "word", y = "count")
    ax.set(ylabel = 'Count')
    plt.show()
  
```

```python
fdist =FreqDist(text)
words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())}) 
```
